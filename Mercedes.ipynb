```python
X=df.iloc[:,2:378]
y=df.iloc[:,1]
from sklearn.model_selection._split import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
print(X_train.head())
print(X_train.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(handle_unknown='ignore')
xcat_train=X_train.iloc[:,0:9]
xcon_train=X_train.iloc[:,9:376]
print(xcat_train.shape)

xcat_test=X_test.iloc[:,0:9]
xcon_test=X_test.iloc[:,9:376]
print(xcat_test.shape)

print(xcon_train.shape)
print(xcon_test.shape)
```

```python
xcat_train.isna().sum().sum()
```

```python
xcat_train=enc.fit_transform(xcat_train)
print(xcat_train.shape)
xcat_test=enc.transform(xcat_test)
print(xcat_test.shape)
```

```python
x3=xcat_train.toarray()
x4=xcat_test.toarray()
x5=pd.DataFrame(x3)
x6=pd.DataFrame(x4)
print(x5.shape)
print(x6.shape)
```

```python
print(x5.isna().sum().sum())
print(x6.isna().sum().sum())
```

```python
xcon_train.isna().sum().sum()
```

```python
x5=np.asmatrix(x5)
```

```python
xcon_train=np.asmatrix(xcon_train)
```

```python
xtrain=np.concatenate((x5,xcon_train), axis=1)
print(xtrain.shape)
```

```python
xtest=np.concatenate((x6,xcon_test), axis=1)
print(xtest.shape)
```

```python
xtrain=pd.DataFrame(xtrain)
xtest=pd.DataFrame(xtest)
print(xtrain.shape)
```

```python
import statsmodels.api as sm
```

```python
y_train=np.asarray(y_train)
y_train=pd.DataFrame(y_train)
print(y_train.shape)
#print(xtrain.shape)
```

```python
print(y_test.shape)
print(xtest.shape)
```

```python
linmodel=sm.OLS(y_train,xtrain)
linresults = linmodel.fit()
print(linresults.summary())
```

```python
lin_pred_train=linresults.predict(xtrain)
print(lin_pred_train)
```

```python
lin_pred_test=linresults.predict(xtest)
print(lin_pred_test)
```

```python
lin_pred_train=np.asarray(lin_pred_train)
lin_pred_train=pd.DataFrame(lin_pred_train)
print(lin_pred_train.shape)
```

```python
print(y_test.shape)
```

```python
from sklearn.metrics import mean_squared_error
train_error=mean_squared_error(y_train,lin_pred_train)
print(train_error)
```

```python
from sklearn.metrics import mean_squared_error
test_error=mean_squared_error(y_test,lin_pred_test)
print(test_error)
```
